```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.spatial.transform import Rotation

m=60
X=np.zeros((m,3))
np.random.seed(42)

angles=(np.random.rand(m)** 3 + 0.5) *2 * np.pi
X[:,0], X[:,1]=np.cos(angles), np.sin(angles) * 0.5
X += 0.28 * np.random.randn(m,3)
X=Rotation.from_rotvec([np.pi / 29, -np.pi /20, np.pi / 4]).apply(X)
X += [0.2, 0, 0.2]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X2D = pca.fit_transform(X)
```

```python
X2D.shape
```

```python
pca.components_      #ayiran xettlere component deyilir, bunlar o vektorlarin valuelaridir
```

```python
pca.explained_variance_ratio_   #her kecirdiyi vektorda ne qeder data oldugun faizle verir, 1cide 75%, 2cide 15%
```

```python
pca.explained_variance_ratio_.sum()
```

```python
from sklearn.datasets import fetch_openml

mnist= fetch_openml('mnist_784', as_frame=False, parser='auto')
X_train, y_train= mnist.data[:60_000], mnist.target[:60_000]
X_test, y_test= mnist.data[60_000:], mnist.target[60_000:]

pca=PCA()
pca.fit(X_train)
cumsum=np.cumsum (pca.explained_variance_ratio_)
d= np.argmax(cumsum >=0.95) + 1
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X_train)
```

```python
pca.n_components_
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc= RandomForestClassifier()
rfc.fit(X_train,y_train)
rfc.score(X_test,y_test)
```

```python
test_reduced=pca.transform(X_test)
```

```python
rfc.fit(X_reduced, y_train)
```

```python
rfc.score(test_reduced, y_test)
```

```python
plt.figure(figsize=(10,7))
plt.plot(cumsum,linewidth=3)
plt.axis([0,400,0,1])
plt.xlabel('Dimensions')
plt.ylabel("Explained varience")
plt.plot([d, d], [0, 0.95], "k:")
plt.plot([0,d], [0.95, 0.95],"k:")
plt.plot(d, 0.95,"ko")
plt.annotate("Elbow", xy=(65, 0.85), xytext = (70, 0.7),
             arrowprops = dict(arrowstyle= "->"))

plt.grid(True)
plt.show()
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import RandomizedSearchCV

pipeline = make_pipeline(PCA(),
                         RandomForestClassifier(random_state=42))
param_grid = {
    "pca__n_components": np.arange(10,80),
    "randomforestclassifier__n_estimators": np.arange(50,500)
}

rscv = RandomizedSearchCV (pipeline, param_grid, n_iter=10, cv=3)
rscv.fit(X_train[:1000], y_train[:1000])
```

```python
print(rscv.best_params_)
```

```python
rscv.best_estimator_.fit(X_train,y_train)
```

```python
rscv.best_estimator_.score(X_train,y_train), rscv.best_estimator_.score(X_test,y_test)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV

sgd_pipeline= make_pipeline(PCA(random_state=42),
                           SGDClassifier(random_state=42))
sgd_param_grid = {
    "pca__n_components": np.arange(10,80)
}

sgd_gscv = GridSearchCV(sgd_pipeline, sgd_param_grid, cv=3)
sgd_gscv.fit(X_train[:1_000], y_train[:1_000])
```

```python
sgd_gscv.best_params_
```

```python
best_model=sgd_gscv.best_estimator_
```

```python
best_model.fit(X_train,y_train)
```

```python
best_model.score(X_train,y_train), best_model.score(X_test,y_test)
```
