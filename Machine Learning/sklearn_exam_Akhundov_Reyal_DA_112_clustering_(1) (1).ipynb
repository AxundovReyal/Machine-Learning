```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score, make_scorer

X = df.copy()


num_features = X.select_dtypes(include=[np.number]).columns
cat_features = X.select_dtypes(exclude=[np.number]).columns

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])


preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ])

pca = PCA(n_components=2)


estimator = KMeans(n_clusters=3, random_state=42, n_init=10)


full_pipeline_with_estimator = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', pca),
    ('estimator', estimator)
])


def silhouette_scorer(X, labels):
    return silhouette_score(X, labels)

silhouette_scorer = make_scorer(silhouette_score, greater_is_better=True)

param_grid = {
    'estimator__n_clusters': [2, 3, 4, 5, 6],
    'estimator__init': ['k-means++', 'random']
}


def custom_silhouette_score(estimator, X, y=None):

    X_transformed = estimator.named_steps['pca'].transform(
        estimator.named_steps['preprocessor'].transform(X)
    )
    labels = estimator.named_steps['estimator'].predict(X_transformed)

    return silhouette_score(X_transformed, labels)


grid_search = GridSearchCV(
    full_pipeline_with_estimator,
    param_grid,
    cv=5,
    scoring=custom_silhouette_score,
    n_jobs=-1,
    error_score='raise'
)

grid_search.fit(X)


print(f"∆èn yax≈üƒ± parametrl…ôr: {grid_search.best_params_}")


best_model = grid_search.best_estimator_

X_transformed = best_model.named_steps['pca'].transform(preprocessor.fit_transform(X))

X['Cluster'] = best_model.named_steps['estimator'].labels_

import matplotlib.pyplot as plt

plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=X['Cluster'], cmap='viridis', alpha=0.7)
plt.title("PCA Sonrasƒ± K-Means K√ºmeleme")
plt.xlabel("Bile≈üen 1")
plt.ylabel("Bile≈üen 2")
plt.colorbar(label="K√ºme Numarasƒ±")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score, make_scorer
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt

X = df.copy()

num_features = X.select_dtypes(include=[np.number]).columns
cat_features = X.select_dtypes(exclude=[np.number]).columns

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ])

pca = PCA(n_components=0.95)

estimator = DBSCAN(eps=0.5, min_samples=5)

full_pipeline_with_estimator = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', pca),
    ('estimator', estimator)
])

def silhouette_scorer(X, labels):
    return silhouette_score(X, labels)

silhouette_scorer = make_scorer(silhouette_score, greater_is_better=True)

param_grid = {
    'estimator__eps': [0.3, 0.5, 0.7],
    'estimator__min_samples': [3, 5, 7]
}

def custom_silhouette_score(estimator, X, y=None):
    X_transformed = estimator.named_steps['pca'].transform(
        estimator.named_steps['preprocessor'].transform(X)
    )
    labels = estimator.named_steps['estimator'].fit_predict(X_transformed)
    return silhouette_score(X_transformed, labels)

grid_search = GridSearchCV(
    full_pipeline_with_estimator,
    param_grid,
    cv=5,
    scoring=custom_silhouette_score,
    n_jobs=-1,
    error_score='raise'
)

grid_search.fit(X)

print(f"∆èn yax≈üƒ± parametrl…ôr: {grid_search.best_params_}")

best_model = grid_search.best_estimator_

X_transformed = best_model.named_steps['pca'].transform(preprocessor.fit_transform(X))

def remove_outliers(X, n_neighbors=10, contamination=0.05):
    neighbors = NearestNeighbors(n_neighbors=n_neighbors)
    neighbors_fit = neighbors.fit(X)
    distances, indices = neighbors_fit.kneighbors(X)
    distance_threshold = np.percentile(distances[:, -1], 100 * (1 - contamination))
    outliers = distances[:, -1] > distance_threshold
    return X[~outliers]

X_cleaned = remove_outliers(X_transformed)

# X_cleaned numpy.ndarray olduƒüu √º√ß√ºn pandas DataFrame-…ô √ßevirm…ôk
X_cleaned_df = pd.DataFrame(X_cleaned, columns=[f'PC{i+1}' for i in range(X_cleaned.shape[1])])

best_model.named_steps['estimator'].fit(X_cleaned_df)

cluster_labels = best_model.named_steps['estimator'].fit_predict(X_cleaned_df)

# Yalnƒ±z t…ômizl…ônmi≈ü veril…ônl…ôr…ô uyƒüun cluster etiketl…ôri …ôlav…ô edirik
X_cleaned_df['Cluster'] = cluster_labels

# Visualizasiyanƒ± d√ºz…ôldirik: X_transformed il…ô cluster etiketl…ôrini uyƒüunla≈üdƒ±rƒ±rƒ±q
plt.scatter(X_transformed[:len(X_cleaned_df), 0], X_transformed[:len(X_cleaned_df), 1], c=X_cleaned_df['Cluster'], cmap='viridis', alpha=0.7)
plt.title("PCA Sonrasƒ± Outlier-l…ôri T…ômizl…ôy…ôn DBSCAN K√ºmeleme")
plt.xlabel("Bile≈üen 1")
plt.ylabel("Bile≈üen 2")
plt.colorbar(label="K√ºme Numarasƒ±")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

labels = best_model.named_steps['estimator'].labels_

score = silhouette_score(X_cleaned_df, labels)
print("Silhouette Score:", score)
```
